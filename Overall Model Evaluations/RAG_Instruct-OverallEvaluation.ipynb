```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rouge_cosine_metrics_gpt4all = pd.read_csv('rag_instruct_benchmark_tester/GPT4ALL/Evaluation/evaluation_summary.csv')
rouge_cosine_metrics_gpt35 = pd.read_csv('rag_instruct_benchmark_tester/GPT_3.5_TURBO/Evaluation/evaluation_summary.csv')
rouge_cosine_metrics_claude = pd.read_csv('rag_instruct_benchmark_tester/Claude/Evaluation/evaluation_summary.csv')
```

```python
plt.figure(figsize=(15, 5))

# Rouge-L Comparison
plt.subplot(1, 2, 1)
plt.plot(rouge_cosine_metrics_gpt4all['NumLabels'], rouge_cosine_metrics_gpt4all['TotalAverageROUGE-L'], label='gpt4all', marker='s', linestyle='--', color='green', linewidth=2)
plt.plot(rouge_cosine_metrics_gpt35['NumLabels'], rouge_cosine_metrics_gpt35['TotalAverageROUGE-L'], label='gpt 3.5', marker='^', linestyle='-', color='orange', linewidth=2)
plt.plot(rouge_cosine_metrics_claude['NumLabels'], rouge_cosine_metrics_claude['TotalAverageROUGE-L'], label='claude', marker='*', linestyle='-', color='red', linewidth=2)


plt.xlabel('Number of Labels')
plt.ylabel('Total Average ROUGE-L Score')
plt.title('After Fine-Tuning ROUGE-L Score Comparison')
plt.legend()

# Cosine Similarity scores
plt.subplot(1, 2, 2)
plt.plot(rouge_cosine_metrics_gpt4all['NumLabels'], rouge_cosine_metrics_gpt4all['TotalAverageCosineSimilarity'], label='gpt4all', marker='s', linestyle='--', color='green', linewidth=2)
plt.plot(rouge_cosine_metrics_gpt35['NumLabels'], rouge_cosine_metrics_gpt35['TotalAverageCosineSimilarity'], label='gpt 3.5', marker='^', linestyle='-', color='orange', linewidth=2)
plt.plot(rouge_cosine_metrics_claude['NumLabels'], rouge_cosine_metrics_claude['TotalAverageCosineSimilarity'], label='Claude', marker='*', linestyle='-', color='red', linewidth=2)


plt.xlabel('Number of Labels')
plt.ylabel('Total Average Cosine Similarity Score')
plt.title('After Fine-Tuning Cosine Similarity Score')
plt.legend()


plt.suptitle('Model Performance Comparison on RAG Instruct Benchmark Tester')


plt.tight_layout()  
plt.show()
```
