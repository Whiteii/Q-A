```python
import nltk
from nltk.util import ngrams
from nltk.metrics import edit_distance
```

```python

def compute_rouge_n(candidate, reference, n):
    candidate_ngrams = list(ngrams(candidate.split(), n))
    reference_ngrams = list(ngrams(reference.split(), n))
    
    candidate_ngrams_count = len(candidate_ngrams)
    reference_ngrams_count = len(reference_ngrams)
    
    overlap_count = sum(1 for ngram in candidate_ngrams if ngram in reference_ngrams)
    
    if reference_ngrams_count == 0:
        return 0.0
    
    return overlap_count / reference_ngrams_count
```

```python
def compute_rouge_l(candidate, reference):
    m, n = len(candidate), len(reference)
    print(m,n)
    dp_table = [[0] * (n + 1) for _ in range(m + 1)]
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if candidate[i - 1] == reference[j - 1]:
                dp_table[i][j] = dp_table[i - 1][j - 1] + 1
            else:
                dp_table[i][j] = max(dp_table[i - 1][j], dp_table[i][j - 1])
    
    return dp_table[m][n] / n
```
